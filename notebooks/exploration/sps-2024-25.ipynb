# Building Per-36 Projections
Using the stats from this to get league average player statistics: https://www.basketball-reference.com/leagues/NBA_stats_totals.html

*Copied from https://www.basketball-reference.com/about/projections.html*

**Give the 2007-08 season a weight of 6, the 2006-07 season a weight of 3, and the 2005-06 season a weight of 1 and calculate the weighted sum of minutes played.** The previous three seasons Howard played 3088, 3023, 3021 minutes, so his weighted sum is 6 * 3088 + 3 * 3023 + 1 * 3021 = 30618 minutes played.


**Use the same weights as above and calculate the weighted sum of rebounds.** The previous three seasons Howard grabbed 1161, 1008, and 1022 rebounds, so his weighted sum is 6 * 1161 + 3 * 1008 + 1 * 1022 = 11012 rebounds.


**Calculate the weighted sum of rebounds for a league average player playing Howard's minutes, then scale this figure to 1000 minutes.** The weighted sum for a league average player playing these minutes is 6 * 3088 * (103271 / 594100) + 3 * 3023 * (100994 / 595750) + 1 * 3021 * (100754 / 595550) = 5269.180 rebounds. Scaled to 1000 minutes this is 1000 * (5269.180 / 30618) = 172.094 rebounds.


**Calculate the player's projected per-36 minute value.** To do this we will find the sum of (2) and (3); divide by the sum of (1) and 1000; and multiply by 36: (11012 + 172.094) / (30618 + 1000) * 36 = 12.734 rebounds per 36 minutes.


**Calculate and apply the age adjustment.** If the player is younger than 28, then the age adjustment is equal to (28 - age) * 0.004. If the player is older than 28, then the adjustment is equal to (28 - age) * 0.002. Howard's 2008-09 season was his age 23 season, so his age adjustment is (28 - 23) * 0.004 = 0.02. Applying this to the figure in (4) we get (1 + 0.02) * 12.734 = 13.0 rebounds per 36 minutes.

```python
# LEAGUE TOTALS:
mp_2022 = 593750
x2p_2022 = 99930 - 30598
x3p_2022 = 30598
ft_2022 = 41657
trb_2022 = 109347
ast_2022 = 60636
stl_2022 = 18772
blk_2022 = 11594
tov_2022 = 33857

mp_2023 = 594850
x2p_2023 = 103260 - 30359
x3p_2023 = 30359
ft_2023 = 45248
trb_2023 = 106858
ast_2023 = 62279
stl_2023 = 17933
blk_2023 = 11448
tov_2023 = 34675

mp_2024 = 593750
x2p_2024 = 103739 - 31579
x3p_2024 = 31579
ft_2024 = 41903
trb_2024 = 107112
ast_2024 = 65610
stl_2024 = 18387
blk_2024 = 12649
tov_2024 = 33468
```

```python

# weighted_minutes = 6 * projections_per_36['mp_2024'] + 3 * projections_per_36['mp_2023'] + projections_per_36['mp_2022']
# weighted_val = 6 * projections_per_36[val_2024] + 3 * projections_per_36[val_2023] + projections_per_36[val_2022]
# league_avg = (1000 / weighted_minutes) * (6 * projections_per_36[mp_2024] * (val_2024/mp_2024) + 3 * projections_per_36[mp_2023] * (val_2023/mp_2023) + projections_per_36[mp_2022] * (val_2022/mp_2022))
# mean_regression = (weighted_val + league_avg) / (weighted_minutes + 1000)
# if projections_per_36['age'] > 28: end_val = (28 - age) * 0.002 * mean_regression
# if projections_per_36['age'] < 28: end_val = (28 - age) * 0.004 * mean_regression

# BUILDING x2p PROJECTIONS
projections_per_36 = player_totals_3_years
projections_per_36['weighted_mp' ] = 6 * projections_per_36['mp_2024'] + 3 *  projections_per_36['mp_2023'] +  projections_per_36['mp_2022']
projections_per_36['weighted_x2p'] = 6 * projections_per_36['x2p_2024'] + 3 * projections_per_36['x2p_2023'] + projections_per_36['x2p_2022']
projections_per_36['weighted_x3p'] = 6 * projections_per_36['x3p_2024'] + 3 * projections_per_36['x3p_2023'] + projections_per_36['x3p_2022']
projections_per_36['weighted_ft' ] = 6 * projections_per_36['ft_2024'] + 3 *  projections_per_36['ft_2023'] +  projections_per_36['ft_2022']
projections_per_36['weighted_trb'] = 6 * projections_per_36['trb_2024'] + 3 * projections_per_36['trb_2023'] + projections_per_36['trb_2022']
projections_per_36['weighted_ast'] = 6 * projections_per_36['ast_2024'] + 3 * projections_per_36['ast_2023'] + projections_per_36['ast_2022']
projections_per_36['weighted_stl'] = 6 * projections_per_36['stl_2024'] + 3 * projections_per_36['stl_2023'] + projections_per_36['stl_2022']
projections_per_36['weighted_blk'] = 6 * projections_per_36['blk_2024'] + 3 * projections_per_36['blk_2023'] + projections_per_36['blk_2022']
projections_per_36['weighted_tov'] = 6 * projections_per_36['tov_2024'] + 3 * projections_per_36['tov_2023'] + projections_per_36['tov_2022']
projections_per_36['proj_mpg'] = (6 * projections_per_36['mpg_2024'] + 3 * projections_per_36['mpg_2023'] + projections_per_36['mpg_2022']) / 10

def calculate_weighted_mpg(row):
    total_weight = 0
    weighted_sum = 0

    # Add weights conditionally based on 'mp' values
    if row['mp_2024'] > 0:
        weighted_sum += 6 * row['mpg_2024']
        total_weight += 6
    if row['mp_2023'] > 0:
        weighted_sum += 3 * row['mpg_2023']
        total_weight += 3
    if row['mp_2022'] > 0:
        weighted_sum += row['mpg_2022']
        total_weight += 1

    # Avoid division by zero if total_weight is 0
    return weighted_sum / total_weight if total_weight > 0 else 0

# Apply the function row by row to calculate 'proj_mpg'
projections_per_36['proj_mpg'] = projections_per_36.apply(calculate_weighted_mpg, axis=1)


def adjust_row(row):
    if row['age'] < 28:
        adjusted_mpg = row['proj_mpg']
        if row['proj_mpg'] < 30:
            adjusted_mpg = np.minimum((row['proj_mpg'] * (1 + ((28 - row['age']) * 0.02))), 36)
    else:
        adjusted_mpg = (row['proj_mpg'] * (1 + ((28 - row['age']) * 0.01)))
    
    return adjusted_mpg

projections_per_36['proj_mpg'] = projections_per_36.apply(adjust_row, axis=1)
```

```python
projections_per_36['proj_x2p'] = (
    (projections_per_36['weighted_x2p'] + (
        (6 * projections_per_36['mp_2024'] * (x2p_2024/mp_2024) +
        3 * projections_per_36['mp_2023'] * (x2p_2023/mp_2023) +
        1 * projections_per_36['mp_2022'] * (x2p_2022/mp_2022)) / projections_per_36['weighted_mp']  *
        (1000 / projections_per_36['mp_2024'])))
    / 
    (projections_per_36['weighted_mp'] + 1000)
    * 36
)

projections_per_36['proj_x2p'] = projections_per_36.apply(lambda row: row['proj_x2p'] * (1 + (28 - row['age']) * 0.002) if row['age'] > 28 
                             else row['proj_x2p'] * (1 + (28 - row['age']) * 0.004), axis=1)

print(projections_per_36[['player', 'proj_x2p', 'mp_2024', 'x2p_2024', 'mp_2023', 'x2p_2023', 'mp_2022', 'x2p_2022']].sort_values(by=['proj_x2p'], ascending=False).head(20))
```
